def initialize_inventory(products):

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(product):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter quantity of {product}: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity can not be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input, Please enter a valid quantity.")

    return inventory
````

````python
def calculate_total_order_value(customer_orders):
    for products in customer_orders:
        Valid_input = False
        while not Valid_input:
            try:
                product_prices = float(input(f"Enter price for {products}: "))
                total_price = 0
                total_price += product_prices

                if product_prices > 0:
                    Valid_input = True
                else:
                    print("Price cannot be 0 or negative. Please enter a valid price.")

            except ValueError:
                print("Invalid input, Please enter a valid price.")
    return total_price
````

````python
def get_customer_order():
    customer_orders = set()
    Valid_input = False
    while not Valid_input:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders > 0:
                Valid_input = True
            else:
                print("Number of orders must be positive. Please enter a valid number.")
        except ValueError:
            print("Invalid input, Please enter a valid number.")

    for _ in range(num_orders):
        while True:
            try:
                product_ordered = input("Which product does customer order?").strip().lower()
                if not product_ordered in products:
                    raise ValueError("Product not found in inventory.")
                customer_orders.add(product_ordered)
                break  
            except ValueError:                        
                print("Product not found in inventory. Please enter a valid product.")
    
    return customer_orders
````
